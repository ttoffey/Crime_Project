```python
import pandas as pd
```

```python
file_path_1 = "crime_state_2014.csv"
file_path_2 = "crime_state_2015.csv"
file_path_3 = "crime_state_2016.csv"
file_path_4 = "crime_state_2017.csv"
file_path_5 = "crime_state_2018.csv"
```

```python
crime_table_2014 = pd.read_csv(file_path_1)
```

```python
crime_table_2015 = pd.read_csv(file_path_2)
```

```python
crime_table_2016 = pd.read_csv(file_path_3)
```

```python
crime_table_2017 = pd.read_csv(file_path_4)
```

```python
crime_table_2018 = pd.read_csv(file_path_5)
```

```python
all_crime_state = pd.concat([crime_table_2014, crime_table_2015,crime_table_2016,crime_table_2017,crime_table_2018])
```

```python
all_crime_state()
```

```python
crime_state_df=all_crime_state.rename(columns={"data_year":"Year","actual":"crime_reported"})
```

```python
del crime_state_df['data_range']

```

```python
del crime_state_df['cleared']
```

```python
del crime_state_df['ori']
```

```python
crime_state_df.head()
```

```python
crime_state_df.reset_index(inplace=True)
crime_state_df.head()
```

```python
crime_state_df.drop(["index", "Unnamed: 0"], inplace=True, axis=1)
```

```python
crime_state_df
```

```python
crime_state_df[["state_abbr","offense","Year",'crime_reported']]
```

```python
all_state_crime=crime_state_df.groupby(['state_abbr','offense'])['crime_reported'].sum().reset_index()
```

```python
all_state_crime.head()
```

```python
coord_df=pd.read_csv("coord.csv")
coord_df=coord_df.rename(columns={"state":"state_abbr","name":"state"})
coord_df.head()
all_state_crime_coord=pd.merge(all_state_crime,coord_df, on="state_abbr",how="inner")
```

```python
all_state_crime_coord.head()
```

```python
all_state_crime_coord.to_csv("all_state_crime.csv")
```
