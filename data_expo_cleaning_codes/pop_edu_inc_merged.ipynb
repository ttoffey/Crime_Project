```python
import pandas as pd
import numpy as np
```

```python
# Population Data Source
# Education Data Source - https://ncses/nsf.gov  - "National Science Foundation"
# Income Data Source - https://fred.stlouis.fed.org - "Federal Reserve Economic Data"
# Crime Data Source - https://crime-data-explorer.app - "FBI"
```

```python
population_df = pd.read_csv("population_final_data.csv")
population_df
```

```python
population_df= population_df.drop(columns=['Unnamed: 0'])
population_df
```

```python
education_df = pd.read_csv("educationData.csv")
education_df
```

```python
education_df.drop(columns=['Unnamed: 0'], inplace=True)
education_df
```

```python
income_df=pd.read_csv('incomeData.csv')
income_df
```

```python
income_df.drop(columns=['Unnamed: 0'], inplace=True)
income_df
```

```python
#merged_df=population_df.merge(income_df, how='left', on='State')
merged_df = population_df.merge(income_df, left_on=['year', 'state'], right_on=['Year', 'State'])
```

```python
merged_df = merged_df.merge(education_df, left_on=['year', 'state'], right_on=['Year', 'State'])
```

```python
merged_df  #Population Income Education
```

```python
merged_df.to_csv('merged_df.csv')
```

```python
merged_df.drop(columns=['Year_x', 'State_x', 'Year_y', 'State_y'], inplace=True)
merged_df
```

```python
police_df=pd.read_csv('state_police.csv')
police_df
```

```python
police_df.drop(columns=['Unnamed: 0'], inplace=True)
police_df
```

```python
merged_df=merged_df.merge(police_df, left_on=['year', 'state'], right_on=['year', 'state_abbr'])
```

```python
merged_df.drop(columns=['state_abbr'], inplace=True)
merged_df
```

```python
merged_df=merged_df.round()
merged_df
```

```python
merged_df.to_csv("PopIncomeEduMerged.csv")
```
